```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Data
credit_df = pd.read_csv('../data/creditcard.csv')
fraud_df = pd.read_csv('../data/Fraud_Data.csv')
ip_df = pd.read_csv('../data/IpAddress_to_Country.csv')
```

```python
# Check data
credit_df.describe()
```

```python
fraud_df.describe()
```

```python
ip_df.describe()
```

```python
# Check for missing values
print("Credit Card")
print(credit_df.isnull().sum())
print("")

print("Fraud Data")
print(fraud_df.isnull().sum())
print("")

print("IP Address to Country")
print(ip_df.isnull().sum())
print("")
```

```python
# Perform EDA
print("\n Credit Card Info")
print(credit_df.info())

print("\n Fraud Data Info")
print(fraud_df.info())

print("\n IP to Country")
print(ip_df.info())
```

```python
fraud_df.signup_time = pd.to_datetime(fraud_df.signup_time)
fraud_df.purchase_time = pd.to_datetime(fraud_df.purchase_time)
```

```python
fraud_df.head()
```

```python
# Visualize 

# Fraud distribution in Fraud_Data.csv
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=fraud_df, palette='coolwarm')
plt.title("Fraudulent vs. Non-Fraudulent Transactions (E-commerce)")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

# Fraud distribution in CreditCard.csv
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=credit_df, palette='coolwarm')
plt.title("Fraudulent vs. Non-Fraudulent Transactions (Bank)")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
# Histograms for Fraud_Data.csv
fraud_df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions - Fraud Data", fontsize=16)
plt.show()

# Histograms for CreditCard.csv
credit_df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions - Credit Card Data", fontsize=16)
plt.show()
```

```python
# Correlation Analysis

# Fraud Data
plt.figure(figsize=(10, 6))
sns.heatmap(fraud_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation - Fraud Data")
plt.show()

# Credit Data
plt.figure(figsize=(12, 8))
sns.heatmap(credit_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation - Credit Card Data")
plt.show()
```
