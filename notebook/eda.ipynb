```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Data
credit_df = pd.read_csv('../data/creditcard.csv')
fraud_df = pd.read_csv('../data/Fraud_Data.csv')
ip_df = pd.read_csv('../data/IpAddress_to_Country.csv')
```

```python
fraud_df.signup_time = pd.to_datetime(fraud_df.signup_time)
fraud_df.purchase_time = pd.to_datetime(fraud_df.purchase_time)
```

```python
fraud_df.head()
```

```python
# Visualize 

# Fraud distribution in Fraud_Data.csv
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=fraud_df, palette='coolwarm')
plt.title("Fraudulent vs. Non-Fraudulent Transactions (E-commerce)")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

# Fraud distribution in CreditCard.csv
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=credit_df, palette='coolwarm')
plt.title("Fraudulent vs. Non-Fraudulent Transactions (Bank)")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
# Histograms for Fraud_Data.csv
fraud_df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions - Fraud Data", fontsize=16)
plt.show()

# Histograms for CreditCard.csv
credit_df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions - Credit Card Data", fontsize=16)
plt.show()
```

```python
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])

# Extract time-based features
fraud_df["hour_of_day"] = fraud_df["purchase_time"].dt.hour
fraud_df["day_of_week"] = fraud_df["purchase_time"].dt.dayofweek

# Plot fraud transactions over hours of the day
plt.figure(figsize=(6, 4))
sns.countplot(x='hour_of_day', hue='class', data=fraud_df, palette='coolwarm')
plt.title("Fraud vs Non-Fraud Transactions Over Hours")
plt.xlabel("Hour of the Day")
plt.ylabel("Transaction Count")
plt.show()
```

```python
# Top devices used in fraud
fraud_transactions = fraud_df[fraud_df["class"] == 1]

plt.figure(figsize=(6, 4))
sns.countplot(y="device_id", data=fraud_transactions, order=fraud_transactions["device_id"].value_counts().index[:10], palette="coolwarm")
plt.title("Top 10 Devices Used for Fraud")
plt.show()
```
